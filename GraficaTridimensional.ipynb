##Importar bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

Definir las ecuaciones del sistema, en este caso son:
##3x + 9y - 10z = 2
##4 x - 6y - 4z = -4
##10x - 2y + 8z = 20

```python
A = np.array([[1, 5, 10], [1, -3, 8], [9, 3, 1]])
B = np.array([24, -4, 20])
```

Debemos resolver el sistema de ecuaciones

```python
solution = np.linalg.solve(A, B)
```

Imprimir la solución

```python
print("La solución del sistema de ecuaciones es: ")
print("x =", solution[0])
print("y =", solution[1])
print("z =", solution[2])
```

Graficar las soluciones

1. plt.figure: Nueva figura en la que se pueden colocar gráficos. Una figura puede contener múltiples gráficos.

2. fig.add_subplot(111, projection='3d'): Esta línea agrega un nuevo subplot a la figura. 111 significa que la figura tiene una fila, una columna y este es el primer subplot. projection='3d' especifica que este subplot será tridimensional.

3. np.linspace(-10, 10, 100): Estas líneas generan arreglos de 100 puntos equiespaciados entre -10 y 10 tanto para x como para y. Estos puntos se utilizarán para crear una cuadrícula de coordenadas.

4. np.meshgrid(x, y): crea una cuadrícula de coordenadas

5. Se calculan los valores de Z para cada ecuación del sistema a partir de las ecuaciones implícitas de los planos

6. plot_surface(X, Y, Z, alpha=0.5): Estas líneas dibujan superficies tridimensionales en el subplot ax. Cada llamada a plot_surface() dibuja una superficie definida por la cuadrícula de coordenadas

7. ax.set_xlabel('x'), ax.set_ylabel('y'), ax.set_zlabel('z'): Estas líneas establecen las etiquetas de los ejes x, y, z

8. Finalmente, esta línea muestra la figura completa con todos los subplots y gráficos.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

Z1 = (-A[0,0]*X - A[0,1]*Y + B[0]) / A[0,2]
Z2 = (-A[1,0]*X - A[1,1]*Y + B[1]) / A[1,2]
Z3 = (-A[2,0]*X - A[2,1]*Y + B[2]) / A[2,2]

ax.plot_surface(X, Y, Z1, alpha=0.5)
ax.plot_surface(X, Y, Z2, alpha=0.5)
ax.plot_surface(X, Y, Z3, alpha=0.5)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
